```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
X = [8,10,7,8,15,8,10,12,10,7]
```

```python
y = [180,200,230,120,470,140,180,310,250,110]
```

```python
X = np.array(X).reshape(-1, 1)
```

```python
y = np.array(y).reshape(-1, 1)
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
# Get the regression equation parameters
intercept = model.intercept_[0]
```

```python
slope = model.coef_[0][0]
```

```python
# Print the regression equation
print(f"Regression equation: y = {intercept:.2f} + {slope:.2f} * X")
```

```python
y_predict = model.predict(X)
```

```python
# Plot the original data and the regression line
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_predict, color='red', label='Regression line')
plt.title('CH9, Q27')
plt.xlabel('Megapixels')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
y_predict = model.predict(X)
```

```python
r_squared = r2_score(y, y_predict)
```
