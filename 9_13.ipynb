```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
# The Data
```

```python
x = [22, 27, 32, 48, 65, 85, 120]
```

```python
y = [9.6, 9.6, 10.1, 11.1, 13.5, 17.7, 25.5]
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
f"Slope: {slope:.4f}, Intercept: {intercept:.4f}"
```

```python
def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))
```

```python
# Scatter Diagram
```

```python
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```
