```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
X = [1580,1572,1352,2224,1556,1435,1438,1089,1941,1698,1539,1364,1979,2183,2096,1400,2372,1752,1386,1163]
```

```python
y = [142500,145000,115000,155900,95000,128000,100000,55000,142000,115000,115000,105000,155000,132000,140000,85000,145000,155000,80000,100000]
```

```python
X = np.array(X).reshape(-1, 1)
```

```python
y = np.array(y).reshape(-1, 1)
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
# Get the regression equation parameters
intercept = model.intercept_[0]
```

```python
slope = model.coef_[0][0]
```

```python
# Print the regression equation
print(f"Regression equation: y = {intercept:.2f} + {slope:.2f} * X")
```

```python
y_predict = model.predict(X)
```

```python
# Residuals are simply the predicted y values minus the real ones, or vice-versa.
residuals = np.array(y) - np.array(y_predict)
```

```python
# Create the residual plot
plt.figure(figsize=(8, 6))
plt.scatter(y_predict, residuals, color='blue', edgecolor='black')
plt.axhline(0, color='red', linestyle='--', linewidth=2)  # Add horizontal line at y=0 (reference line)
plt.title("Residual Plot", fontsize=16)
plt.xlabel("Predicted Values", fontsize=14)
plt.ylabel("Residuals", fontsize=14)
plt.grid(True)
plt.show()
```
